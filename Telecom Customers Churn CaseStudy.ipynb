## Exploratory Data Analysis

    #### Univariate Analysis
        - Cateogrical 
        - Continous
    
    
    #### Bivariate Analysis

```python
df.head(2) 
```

```python
# Partitioning the Data based on our Target Variable 'churn'

df_1 = df[df['Churn']=='Yes'] ## dataframe containing churned customers
df_0 = df[df['Churn']=='No'] ## dataframe contianing non-churned customers
```

```python
df_1.head(3)
```

```python
df_0.head(3)
```

```python
# Let us begin with Analysis process
import warnings
warnings.filterwarnings('ignore')
```

#### Univariate Analysis (Continous)
    - MonthlyCharges

```python
# let us use distplot from seaborn

sns.distplot(df_1['MonthlyCharges'],hist=False,label='Churn')
sns.distplot(df_0['MonthlyCharges'],hist=False,label='Non-Churn')
plt.legend()
plt.show()
```

### Univariate Analysis(Categorical)

    - Contract

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title('Churn')
sns.countplot(x='Contract',data=df_1)

plt.subplot(1,2,2)
plt.title('Non-Churn')
sns.countplot(x='Contract',data=df_0)
plt.show()
```

```python
## Bivariate Analysis  (Conti-Conti)

# Tenure + MonthlyCharges

df.head(3)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title('Churn')
plt.scatter(df_1['MonthlyCharges'],df_1['tenure'])

plt.subplot(1,2,2)
plt.title('Non-Churn')
plt.scatter(df_0['MonthlyCharges'],df_0['tenure'])
plt.show()
```

```python
# Bivariate Cate-Continous

#  let us analyse the monthly spending behaviour of customers based on their contracttype.

df.head(3)
```

```python

#  let us analyse the monthly spending behaviour of customers based on their contracttype.
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title('Churn')
sns.boxplot(x='Contract',y='MonthlyCharges',data=df_1)
plt.subplot(1,2,2)
plt.title('Non-Churn')

sns.boxplot(x='Contract',y='MonthlyCharges',data=df_0)
plt.show()
```

#### Bivariate Cate-Cate 

    - Contract , Gender

```python
# let us analyse if gender plays any role in customer churn
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title('Churn')

sns.countplot(x='Contract',hue='gender',data=df_1,palette='Set1')

plt.subplot(1,2,2)
plt.title('Non-Churn')

sns.countplot(x='Contract',hue='gender',data=df_0)


plt.show()

# let us analyse if gender plays any role in customer churn
```

```python
# let us observe the distribution of MonthlyCharges in data
# (binning)
```

```python
pd.cut(df['MonthlyCharges'],bins=5,labels=['Below Avg','Mean','Above Avg','Good','Better']).value_counts().plot(kind='bar')
plt.show()
```
